```python
wine_new=wine.drop('total sulfur dioxide',axis=1)
wine_new.head()
```

```python
wine_ml=pd.get_dummies(wine_new,drop_first=True)
wine_ml.tail()
```

```python
wine_ml.isnull().sum()
```

```python
Y=wine_ml['quality'].apply(lambda y: 1 if y>7 else 0)
Y
```

```python
X=wine_ml.drop('quality',axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
scaler.fit(X)
X_standard=scaler.transform(X)
```

```python
X=X_standard
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=.2,random_state=245)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logreg=LogisticRegression()
logreg.fit(X_train,Y_train)
```

```python
y_pred=logreg.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
accuracy_score(Y_test,y_pred)
```

```python
print(classification_report(Y_test,y_pred))
```

```python
print(confusion_matrix(Y_test,y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier()
rfc.fit(X_train,Y_train)
```

```python
rfc_pred=rfc.predict(X_test)
```

```python
accuracy_score(Y_test,rfc_pred)
```

```python
print(classification_report(Y_test,rfc_pred))
```

```python
print(confusion_matrix(Y_test,rfc_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc=DecisionTreeClassifier()
dtc.fit(X_train,Y_train)
```

```python
dtc_pred=dtc.predict(X_test)
```

```python
accuracy_score(Y_test,dtc_pred)
```

```python
print(classification_report(Y_test,dtc_pred))
```

```python
print(confusion_matrix(Y_test,dtc_pred))
```

```python
import xgboost as xgb

# Assuming X_train and Y_train are your training data
xgb_classifier = xgb.XGBClassifier()
xgb_classifier.fit(X_train, Y_train)
```

```python
xgb_pred=xgb_classifier.predict(X_test)
```

```python
accuracy_score(Y_test,xgb_pred)
```

```python
print(classification_report(Y_test,xgb_pred))
```

```python
print(confusion_matrix(Y_test,xgb_pred))
```

```python
models=logreg,rfc,dtc,xgb_classifier
model_names=['logreg','rfc','dtc','xgb_classifier']
accuracy_scores = [accuracy_score(Y_test, model.predict(X_test)) for model in models]
```

```python
plt.bar(model_names, accuracy_scores, color=['blue', 'green', 'orange', 'red'])
plt.xlabel('Models')
plt.ylabel('Accuracy Score')
plt.title('Accuracy Scores of Different Models')
plt.ylim([0, 1])
plt.show()
```
